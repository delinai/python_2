```python
def fix_phone(x):
    pattern1 = re.compile(r'^(\d{3})-(\d{3})-(\d{4})$')
    pattern2 = re.compile(r'^(\d{10})$')
    pattern3 = re.compile(r'^(\d{6})-(\d{4})$')
    pattern4 = re.compile(r'^(\d{3})-(\d{7})$')

    if pattern1.search(x) != None:
        return x
    elif pattern2.search(x) != None:
        return '-'.join([x[:3],x[3:6],x[6:]])
    elif pattern3.search(x) != None:
        return '-'.join([x[:3],x[3:]])
    elif pattern4.search(x) != None:
        return '-'.join([x[:7],x[7:]])

    else:
        return "pattern not found"
```

```python
df['PhoneNumber'] = df['PhoneNumber'].apply(fix_phone)
```
