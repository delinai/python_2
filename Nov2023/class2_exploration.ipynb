```python
df.groupby('bedrooms')['price'].count()
```

```python
df_fam = df[(df['bedrooms']>=2) & (df['bedrooms']<=5)]
```

```python
df_fam.groupby('bathrooms')['price'].count()
```

```python
df_fam.groupby('waterfront')['price'].count()
```

```python
df_fam.groupby('view')['price'].count()
```

```python
df_fam.columns
```

```python
# bivariate - continuous
sns.scatterplot(x=df_fam['sqft_living'],y=df_fam['price'])
```

```python
sns.scatterplot(x=df_fam['sqft_lot'],y=df_fam['price'])
```

```python
sns.boxplot(x='bedrooms', y='price', data=df_fam)
```

```python
sns.boxplot(x='waterfront', y='price', data=df_fam)
```

```python
sns.boxplot(x='condition', y='price', data=df_fam)
```

```python
sns.boxplot(x='grade', y='price', data=df_fam)
```

```python
sns.scatterplot(x=df_fam['sqft_living'], y=df_fam['price'], hue=df_fam['grade'])
```

```python
df_fam.groupby(['bedrooms','grade'])['price'].median().unstack('grade').plot(kind='bar', stacked=False)
```

```python
df_fam.groupby(['waterfront','grade'])['price'].median().unstack('grade').plot(kind='bar', stacked=False)
```

```python
df_fam.groupby(['waterfront','bedrooms'])['price'].median().unstack('bedrooms').plot(kind='bar', stacked=False)
```

```python
sns.scatterplot(x=df_fam[df_fam['waterfront']==1]['sqft_living'], y=df_fam[df_fam['waterfront']==1]['price'], hue=df_fam[df_fam['waterfront']==1]['bedrooms'])
```

```python
sns.scatterplot(x=df_fam[df_fam['waterfront']==1]['sqft_living'], y=df_fam[df_fam['waterfront']==1]['price'], hue=df_fam[df_fam['waterfront']==1]['bedrooms'], size=df_fam[df_fam['waterfront']==1]['sqft_lot'])
```
