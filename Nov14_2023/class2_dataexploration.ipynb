```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/delinai/schulich_ds1/main/Datasets/kc_house_data.csv')
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.loc[:,:'view'].describe(include='all')
```

```python
df.loc[:,'view':].describe(include='all')
```

```python
df.groupby('date')['id'].count().sort_index().plot()
```

```python
df.groupby('date')['id'].count().describe()
```

```python
df.groupby('date')['price'].median().sort_index().plot()
```

```python
# price distribution
sns.histplot(df['price']/1000)
```

```python
# focus on average homes
df_avg = df[df['price']<=750000]
```

```python
sns.histplot(df_avg['price'])
```

```python
sns.histplot(df_avg['sqft_living'])
```

```python
# categorical univariate
df_avg.groupby('bedrooms')['id'].count()/len(df_avg)
```

```python
# focus on 2-4 bedroom homes
df_fam = df_avg[df_avg['bedrooms'].isin([2,3,4])]
```

```python
df_fam.groupby('bedrooms')['id'].count()
```

```python
# check the number of bathrooms
df_fam.groupby('bathrooms')['id'].count()/len(df_fam)
```

```python
# waterfront
df_fam.groupby('waterfront')['id'].count()/len(df_fam)
```

```python
# view
df_fam.groupby('view')['id'].count()/len(df_fam)
```

```python
# bivariate
sns.scatterplot(x=df_fam['sqft_living'], y=df_fam['price'])
```
